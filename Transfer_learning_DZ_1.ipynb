## 1) Скачать нейросети ResNet и написать короткую процедуру для предсказания класса изображения

```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import requests
from io import BytesIO
```

```python
# Загрузка предварительно обученной модели ResNet50
model = ResNet50(weights='imagenet') # Программа использует модель ResNet50 для предсказания класса изображения
```

```python
def predict_image_class(url):
    response = requests.get(url)
    # предобработка изображения
    img = image.load_img(BytesIO(response.content), target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    # предсказание класса изображения
    predictions = model.predict(img_array)
    decoded_predictions = decode_predictions(predictions, top=3)[0] # выводим топ-3 предсказанных класса с соответствующими вероятностями.

    print("Predictions:")
    for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
        print(f"{i + 1}: {label} ({score:.2f})")
```

```python
# Пример URL изображения
image_url = "https://klike.net/uploads/posts/2023-03/1678950286_1.jpg"
```

```python
# Запуск программы
predict_image_class(image_url)
```

## 2) Скачать нейросеть BERT для лингвистических задач и реализовать процедуру классификации текстов (без оглядки на качество классификации)

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
```

```python
# Загрузка предварительно обученной модели BERT и токенизатора
model_name = 'bert-base-uncased' # загружаем предварительно обученную модель BERT (bert-base-uncased)
tokenizer = BertTokenizer.from_pretrained(model_name) # загружаем токенизатор
model = BertForSequenceClassification.from_pretrained(model_name) # создаем модель
```

```python
def classify_text(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    outputs = model(**inputs)
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()

    return predicted_class
```

```python
# Пример текста для классификации
# text_to_classify = "This is a sample text for classification." # 1
text_to_classify = "Image dehazing is a crucial low-level vision task that recovers clear images from hazy ones." # 0
```

```python
# Запуск программы
predicted_class = classify_text(text_to_classify)
print(f"The predicted class for the text is: {predicted_class}")
```
